```python
# Data yang diketahui
l40 = 100000
q40_select = 0.0009
q40_plus1 = 0.0012
q42 = 0.0020  # ultimate (tidak dipakai di langkah ini)

# Perhitungan
l40_plus1 = l40 * (1 - q40_select)
l40_plus2 = l40_plus1 * (1 - q40_plus1)

# Ultimate
l42 = l40_plus2

print("a. l[40]+1 =", l40_plus1)
print("   l[40]+2 =", l40_plus2)
print("b. l42 (ultimate) =", l42)
```

```python
# Data
l60 = 10000
q60_select = 0.004
q60_plus1 = 0.005
q62 = 0.007

# Perhitungan l
l60_plus1 = l60 * (1 - q60_select)
l60_plus2 = l60_plus1 * (1 - q60_plus1)

# Probabilitas hidup 3 tahun
p60_3 = (1 - q60_select) * (1 - q60_plus1) * (1 - q62)

print("a. l[60]+1 =", l60_plus1)
print("   l[60]+2 =", l60_plus2)
print("b. Probabilitas seseorang yang terpilih pada usia 60 akan hidup hingga usia 63 =", p60_3)
```
